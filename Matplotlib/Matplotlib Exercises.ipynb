## Exercise 1

1. Create a figure objects called fig
2. Add axis to the figure at [0,0,1,1]. Call this new axis ax.
3. Plot (x,y) on that axes and set the title to `title`, set the y axis label to `y` and set the x axis label to `x`

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

ax.plot(x,y)
```

## Exercise 3
Add two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Plot**

1. Set title of inner axes to `zoom`
2. Set xlimits of inner axes to 20, 22
3. Set ylimits of inner axes to 30, 50
4. Plot innner and outer axes with x and y
5. Set xlabel of both axis to 'X' and set ylabel of inner axes to 'Y' and set ylabel of outer axis to 'Z'

```python
ax1.set_ylabel('Z')
ax1.set_ylabel('X')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20, 22])
ax2.set_ylim([30,50])

ax1.plot(x, x**2)
ax2.plot(x, y)

fig
```

## Exercise 4
Use subplots to create a figure with 2 columns and 1 row

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Plot (x,y) and (x, x ^ 2) on the axes. Set the left plot to dashed lines and set the thickness of both lines to thicker**

```python
axes[0].plot(x,y, linestyle='--', linewidth=4, color='blue')

axes[1].plot(x, x**2, linewidth=4, color='red')

fig
```

**Resize the plot**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,2))

axes[0].plot(x,y, color='blue', linewidth=4)

axes[1].plot(x, x**2, linewidth=4, color='red', linestyle='--')
```
