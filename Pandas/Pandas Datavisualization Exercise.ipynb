```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.plot.scatter(x='a', y='b', figsize=(12,4))
```

**Create histogram of 'a' column**

```python
df3['a'].plot.hist()
```

**Boxplot comparing a and b column**

```python
df3[['a', 'b']].plot.box()
```

**Create kde plot of 'd' column**

```python
df3['d'].plot.kde()
```

**Create an area plot of all the columns for just the rows up to 30**

```python
df3.iloc[0:30].plot.area()
```

```python
df3.iloc[0:30].plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
